```python
import random
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Function To Print random numbers
list1 = [] #EmptyList
def Rtemp():
    r_temp = random.uniform(25, 30)
    list1.append(r_temp)   
```

```python
list2 = []
def Rhumid():
    r_humid = random.uniform(65, 72)
    list2.append(r_humid)
```

```python
list3 = []
def Rsm():
    r_sm = random.uniform(55, 82)
    list3.append(r_sm)
```

```python
for i in range(0,237):
    Rtemp()
    
print(list1)
```

```python
for i in range(0,237):
    Rhumid()
    
print(list2)
```

```python
for i in range(0,237):
    Rsm()
    
print(list3)
```

```python
#PrintingTimeSeries
datetime = pd.date_range('2020-07-01 06:00:00', '2020-07-30 18:00:00', freq='3H')
datetime
```

```python
excelfile = 'RandomSet.xlsx'
df = pd.read_excel(excelfile)
df.head()
```

```python
df['DateAndTime'] = datetime
df['Temp'] = list1
df['Humidity'] = list2
```

```python
df.set_index('DateAndTime')
# df.index = df['DateAndTime']
df.info()
```

```python
writer = pd.ExcelWriter('output.xlsx')
df.to_excel(writer)
writer.save()
print('DataFrame is written successfully to Excel File.')
```
